## 柱状图绘制

```python
names = ["a类型", "b类型", "c类型"]
values = [1, 10, 100]
```

```python
plt.bar(range(len(names)), values)   # 不接受字符串
plt.xticks(range(len(names)), names)
```

```python
plt.scatter(range(len(names)), values)   # scatter也不接受字符串
plt.xticks(range(len(names)), names)
```

```python
plt.scatter(range(len(names)), values)   # scatter也不接受字符串
plt.xticks(range(len(names)), names)
plt.title("离散数据散点图")
```

## 子图绘制--SubPlot

1.将画布切分

2.将plot图分配到固定的位置和大小

```python
plt.figure(1, figsize = (15, 5))

plt.subplot(131)  # rcn
plt.bar(range(len(names)), values, color="r")
plt.xticks(range(len(names)), names)

plt.subplot(132)
plt.scatter(range(len(names)), values, color="y")
plt.xticks(range(len(names)), names)

plt.subplot(133)
plt.plot(range(len(names)), values)
plt.xticks(range(len(names)), names)

plt.suptitle("离散数据的绘图")
```

## Seaborn 绘图练习

基于Pyplot之上提供了更好的数据支持

自带数据集/iris, titanic ...

图形绘制进行了优化

提供了多种可供参考的高级绘图--带状图，蜂群图，箱型图，提琴图，分布图

```python
import seaborn as sns
```

```python
# 加载一个数据集，小费

tips = sns.load_dataset("tips")
tips.head()
```

### 带状图--离散数据和连续数据之间的关系

```python
sns.stripplot(data=tips, x="day", y="total_bill", jitter=False)  # jitter 加横向抖动
```

```python
# jitter 加横向抖动
sns.stripplot(data=tips, x="day", y="total_bill", jitter=True)  
```

### 蜂群图

```python
sns.swarmplot(x="day", y="total_bill", data=tips)
```

```python
tips.head()
```

### Hue--分组参数

```python
sns.swarmplot(x="day", y="total_bill", data=tips, hue="time")
```

```python
sns.swarmplot(x="day", y="total_bill", data=tips, hue="sex")
```

```python
sns.swarmplot(x="day", y="total_bill", data=tips, hue="size")
```

```python
sns.swarmplot(x="size", y="total_bill", data=tips)
```

```python
tips["size"].corr(tips["total_bill"]) # 计算相关系数
```

### 箱线图

```python
sns.boxplot("day", "total_bill", data=tips)
```

```python
sns.boxplot("day", "total_bill", data=tips, hue="time")
```

#### 离群点怎么找出来的

```python
p = np.percentile(tips["total_bill"], 99.5)
```

```python
tips["total_bill"].sort_values(ascending = False).head()
```

```python
tips[tips["total_bill"] > p]   # 离群点删掉或者改成 p
```

### 提琴图 violinplot()

```python
sns.violinplot(x="day", y="total_bill", data=tips)
```

```python
sns.violinplot(x="day", y="total_bill", data=tips, hue="time")
```

```python
sns.violinplot(x="day", y="total_bill", data=tips, hue="time", split=True)
```

### 多图合成一个图

```python
sns.violinplot(x="day", y="total_bill", data=tips)
sns.swarmplot(x="day", y="total_bill", data=tips, color="w")
```
